# Back Propagation

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
traindata=pd.read_excel('ANNdataset (1).xlsx', sheet_name='Train')
```

```python
traindata=traindata.dropna()
from sklearn.model_selection import train_test_split
train,test= train_test_split(traindata,test_size=0.3)
```

```python
train_op = train['output']
train_ip = train.drop(['output','education'],axis = 1)
```

```python
test_op = test['output']
test_ip = test.drop(['output','education'],axis = 1)
```

```python
train_ip.shape
```

```python
train_nn_ip = pd.get_dummies(train_ip)
test_nn_ip = pd.get_dummies(test_ip)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(train_nn_ip)
X_test = sc.transform(test_nn_ip)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras import optimizers
from keras.models import Sequential
from keras.layers import Dense,Activation
from keras.optimizers import SGD
model1=keras.Sequential([keras.layers.Dense(174,input_dim=87,activation=tf.nn.sigmoid),keras.layers.Dense(1),])

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.0, nesterov=True)
model1.compile(optimizer='sgd',loss='mse',metrics=['accuracy'])
model1.fit(X_train,train_op,validation_data=(X_test,test_op),epochs=250)
```

# SVM Classifier

```python
import numpy as np
from sklearn.svm import SVC
clf = SVC(gamma=1,C=100)
clf.fit(X_train, train_op) 
```

```python
predicted_op = clf.predict(X_test)
output = [int(i) for i in predicted_op]
compare_op=[i for i, j in zip(output, test_op) if i == j]
accuracy = (len(compare_op)/len(test_op))* 100
```

```python
accuracy
```

# LVQ2 Classifier

```python
import neupy.algorithms as nyp
```

```python
lvqnet = nyp.LVQ2(
        # number of features
        n_inputs=87,

        # number of classes
        n_classes=2,

        verbose=True,
        show_epoch=20,

        step=0.25,
        n_updates_to_stepdrop=150 * 100,
    )
```

```python
lvqnet.train(X_train, train_op, epochs=100)
```

```python
y=lvqnet.predict(X_test)
```

```python
y
```

```python
compare_op_lvq=[i for i, j in zip(y, test_op) if i == j]
accuracy = (len(compare_op_lvq)/len(test_op))* 100
```

```python
accuracy
```
